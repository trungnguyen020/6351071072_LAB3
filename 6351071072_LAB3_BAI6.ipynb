```python
print("Nguyễn Đức Trung - 6351071072")

import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

try:
    img = plt.imread('dog.png')

    if img.shape[2] == 4:
        img = img[:, :, :3]

    h, w, d = img.shape

    if img.max() > 1.0:
        img = img / 255.0

    img_flat = img.reshape(h * w, d)

    kmeans = KMeans(n_clusters=32, random_state=42, n_init=10)
    kmeans.fit(img_flat)

    new_colors = kmeans.cluster_centers_
    labels = kmeans.labels_

    img_compressed = new_colors[labels].reshape(h, w, d)

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Original')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(img_compressed)
    plt.title('K-Means (k=32)')
    plt.axis('off')

    plt.show()

    # Clip the values to be within [0, 1] before saving
    img_compressed = np.clip(img_compressed, 0, 1)
    plt.imsave('dog_compressed.png', img_compressed)
    print("Done.")

except FileNotFoundError:
    print("File dog.png not found.")
```
