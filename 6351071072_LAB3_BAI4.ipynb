```python
# --- CELL 1: IMPORT VÀ CHUẨN BỊ DỮ LIỆU ---
print("Nguyễn Đức Trung - 6351071072")

import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
iris = load_iris()
X = iris.data
y = iris.target
max_values = X.max(axis=0)
X = X / max_values

print("Đã load và chuẩn hóa dữ liệu Iris.")
print(f"Kích thước dữ liệu: {X.shape}")
```

```python
# --- CELL 2: CÂU C (Định nghĩa Model) ---
print("Nguyễn Đức Trung - 6351071072")

def create_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(4,)))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(3, activation='softmax'))

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

print("Đã định nghĩa hàm create_model thành công.")
```

```python
# --- CELL 3: CÂU D (Huấn luyện K-Fold) ---
print("Nguyễn Đức Trung - 6351071072")

k = 10
kfold = KFold(n_splits=k, shuffle=True, random_state=42)
fold_accuracy = []
fold_precision = []
fold_recall = []
fold_f1 = []
class_precision_list = []
class_recall_list = []
class_f1_list = []

models = []

fold_idx = 1

for train_index, test_index in kfold.split(X):
    X_train_fold, X_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    model = create_model()
    model.fit(X_train_fold, y_train_fold, epochs=100, batch_size=1, verbose=0)
    y_pred_prob = model.predict(X_test_fold, verbose=0)
    y_pred = np.argmax(y_pred_prob, axis=1)
    acc = accuracy_score(y_test_fold, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_test_fold, y_pred, average='weighted', zero_division=0)
    fold_accuracy.append(acc)
    fold_precision.append(prec)
    fold_recall.append(rec)
    fold_f1.append(f1)
    prec_cls, rec_cls, f1_cls, _ = precision_recall_fscore_support(y_test_fold, y_pred, average=None, labels=[0,1,2], zero_division=0)
    class_precision_list.append(prec_cls)
    class_recall_list.append(rec_cls)
    class_f1_list.append(f1_cls)
    models.append(model)

    print(f"Fold {fold_idx}/{k} - Acc: {acc:.4f} - Prec: {prec:.4f} - Rec: {rec:.4f} - F1: {f1:.4f}")
    fold_idx += 1
```

```python
# --- CELL 4: CÂU E (Average Metrics Model) ---
print("Nguyễn Đức Trung - 6351071072")

print("=== KẾT QUẢ TRUNG BÌNH QUA 10 FOLDS (MODEL) ===")
print(f"Average Accuracy : {np.mean(fold_accuracy):.4f}")
print(f"Average Precision: {np.mean(fold_precision):.4f}")
print(f"Average Recall   : {np.mean(fold_recall):.4f}")
print(f"Average F1-score : {np.mean(fold_f1):.4f}")
```

```python
# --- CELL 5: CÂU F (Average Metrics Per Class) ---
print("Nguyễn Đức Trung - 6351071072")

# Chuyển list thành numpy array để tính mean theo cột (axis=0)
avg_prec_cls = np.mean(class_precision_list, axis=0)
avg_rec_cls = np.mean(class_recall_list, axis=0)
avg_f1_cls = np.mean(class_f1_list, axis=0)

target_names = iris.target_names

print("=== KẾT QUẢ TRUNG BÌNH CHO TỪNG CLASS ===")
for i, name in enumerate(target_names):
    print(f"--- Class: {name} ---")
    print(f"  Avg Precision: {avg_prec_cls[i]:.4f}")
    print(f"  Avg Recall   : {avg_rec_cls[i]:.4f}")
    print(f"  Avg F1-score : {avg_f1_cls[i]:.4f}")
```

```python
# --- CELL 6: CÂU G (Dự đoán mẫu mới) ---
print("Nguyễn Đức Trung - 6351071072")
best_fold_index = np.argmax(fold_f1)
best_model = models[best_fold_index]

print(f"Sử dụng Model từ Fold thứ {best_fold_index + 1} (F1-score: {fold_f1[best_fold_index]:.4f})")
new_samples = np.array([
    [6.2, 2.9, 4.3, 1.3],
    [5.1, 3.5, 1.4, 0.2],
    [7.3, 2.8, 6.4, 2.1]
])
new_samples_norm = new_samples / max_values
pred_probs = best_model.predict(new_samples_norm)
pred_labels = np.argmax(pred_probs, axis=1)

print("\n=== KẾT QUẢ DỰ ĐOÁN ===")
for i, sample in enumerate(new_samples):
    class_name = iris.target_names[pred_labels[i]]
    print(f"Mẫu {sample} -> Dự đoán: {class_name}")
```
